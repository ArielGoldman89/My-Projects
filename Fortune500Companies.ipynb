# 5. Top 5 Companies by Revenue for each Decade

**Checking the unique years present in the dataset, ensuring that we have a clear understanding of the decades to be plotted from 1955 to 2005.**

```python
# Identifying unique years
df['Year'].unique()
```

```python
# Creating a subset to show the information
years = df[['Year', 'Revenue (in millions)']]
years
```

* **These lines of code allow an easy interpretation by dividing the years into decades (10 years)**
* **Then I created a subset for decades between 1950 to 2009**

```python
# By decade (10)
period_length = 10

# Specifying range of years
start_year = 1950
end_year = 2009

# Creating subset for decades between 1950-2009
decades = df[(df['Year'] >= 1950) & (df['Year'] <= 2009)]
```

**Using modulo to determine the starting and ending years for the last period and created bins for decades**

```python
year_range = end_year - start_year
modulo = year_range % period_length

print('For 1950–2009 the modulo is:',modulo)
```

**Finding the starting and ending years for the last period**

```python
if modulo == 0:
    final_start = end_year - period_length
else:
    final_start = end_year - modulo
final_end = end_year + 1 # The addition of one is done to include the last year as well.
```

**Creating a list of all the earlier starting years for the year range.**


```python
# Creating a list of all the earlier starting years for the year range.
starts = np.arange(start_year, final_start, period_length).tolist()
```

```python
# Creating lists of tuples
tuples = [(start, start + period_length) for start in starts]

# Adding the last period calculated earlier
tuples.append(tuple([final_start, final_end]))

# Creating the bins 
bins = pd.IntervalIndex.from_tuples(tuples, closed='left')

print('Invervals:\n\n',bins)
print('\nTuples:')
tuples
```

* **Converting bins to strings**
* **Removed the square brakets or parenthesis, then split the label in to two parts using comma and space slecting the first part, converting the second part to an integer.**
* **Zip funcion() I used it to combine original_labels and new_labels into a single iterable.**


```python
# Converting bins to strings
original_labels = list(bins.astype(str))

new_labels = ['{} - {}'.format(b.strip('[)').split(', ')[0], int(b.strip('[)').split(', ')[1])-1)
              for b in original_labels]

label_dict = dict(zip(original_labels, new_labels))

label_dict
```

* **pd.cut function() I used it to segment years and sort data values into bins then converted Decade to string format to ensure compatibility with other string**

```python
# Assigning each row to a decade

df['Decade'] = pd.cut(df['Year'], bins=bins, include_lowest=True, precision=0)
df['Decade'] = df['Decade'].astype("str")
df = df.replace(label_dict)

df.head()
```

* **By grouping Decades with Revenue will allow me to perform the plot asked**

```python
# Grouping Companies by Decades and Revenue

columns = ['Decade', 'Company']

grouped_revenue = df.groupby(columns)['Revenue (in millions)'].sum().reset_index()

grouped_revenue.head()
```

* **Grouped the columns by Decade, Company and Revenue, then applied sum() to aggregate The revenue each Decade. Finally, I sorted Decades and Revenue in descending order and used head() to have the top five highest Revenue for readability.** 

```python
# Grouping the Top 5 Companies by Revenue per Decade

top5_revenue_per_decade = grouped_revenue.groupby(
    ['Decade','Company','Revenue (in millions)']).sum().sort_values(by=['Decade','Revenue (in millions)'],
                                                                ascending=False).groupby('Decade').head().reset_index()
top5_revenue_per_decade
```

```python
# Bar Plot

# Plot size
plt.figure(figsize=(12, 9))

# Executing the Plot
ax = sns.barplot(data=top5_revenue_per_decade, x='Decade', y='Revenue (in millions)',
            hue='Company', palette='tab10')

plt.title('Top 5 Companies by Revenue per Decade',fontsize=20)
plt.xlabel('Decade',fontsize=18)
plt.ylabel('Revenue (in millions)',fontsize=18)

plt.grid(axis='x', linestyle='--', alpha=0.9)
plt.legend(title='Companies', loc='upper left', bbox_to_anchor=(1, 1))

# The x-axis labels were inverted to read the decades from left to right
ax.invert_xaxis()

plt.show()
```

### 5.1. Plot Explanation

This Bar Plot represents the Revenue (in millions) of the top five companies for each decade from 1950 to 2009. The decision to use a Bar Chart for visualizing the Top 5 Companies by Revenue over each decade to compare categorical data (companies) is because it can be divided into groups or categories. By plotting a Bar Chart, the magnitude of the different categories can be clearly distinguished over each decade, as well as the clear separation between each decade using multiple colors for each company, providing a straightforward representation of the Revenue through the decades.


### 5.2. Plot Description

* General Motors (green bar) was the top company in Revenue for every decade from 1950 to 1990. However, it was overtaken by Wal-Mart Stores (blue bar) in the 2000-2009 decade. Surprisingly, Wal-Mart Stores made its first appearance since 1950 in the 2000s.

* Exxon Mobil (orange bar) was the top company in Revenue in the decades 1980-1989 and 1990-1999, and its Revenue has been increasing since the 1950-1959 decade and it has been in the top 5 for every decade since then.

* Ford Motor (red bar) was in the top 5 for every decade from 1950 to 1990, and its Revenue has been increasing since then, however, it went down in the 2000-2009 decade.

* General Electric (purple bar) was in the top 5 for every decade from 1950 to 1960, but it dropped from the top 5 Revenue in the decades 1970-1979 and 1980-1989 then went back to the top 5 since decade 1990-1999.

* IBM (brown bar) was only in the top 5  in the decade 1990-1999, but it dropped out of the top 5 in the 1990s and has not been back since then.

* Mobil (pink bar) was only in the top 5 in the decades 1960-1969, 1970-1979, 1980-1989.

* Texaco (grey bar) was only in the top 5 in the decades 1970-1979, 1980-1989 and has the lowest Revenue compared to its competitors in both decades.

* U.S Steel (light green bar) has only shown in the decade 1950-1959 with low revenue.


**It is interesting to observe how the Revenue of these companies  has increased significantly over the decades. For instance, General Motors had a Revenue of $53.575 in the 50s, while the Revenue for the same company in the 90s was nearly 1.5 million.**

### 5.3. Plot Conclusion

The Bar Plot shows that the top 5 companies by revenue have fluctuated over decades. A number of factors can affect the Revenue such as, technological changes, economic changes and government regulations.

# 6. Top 20 Companies by Revenue in 1971

```python
# Adding relevant columns to the analysis
relevant_columns = ['Year', 'Rank', 'Company', 'Revenue (in millions)']

# In this case 'nlargest' retrives the top 20 largest by filtering the Revenue in 1971
top20_1971 = df[df['Year'] == 1971][relevant_columns].nlargest(20, 'Revenue (in millions)')

# Displaying the results
top20_1971
```

```python
# Horizontal Bar Plot

# top 20 companies based on Revenue in 1971
top20_companies1971 = df[df['Year'] == 1971].nlargest(20, 'Revenue (in millions)')

companies = top20_companies1971['Company']
revenue = top20_companies1971['Revenue (in millions)']

# Plot size
plt.figure(figsize=(10, 8))
# Executing the Plot
plt.barh(companies, revenue, color='#EDB120')

# The y-axis labels was inverted to read the companies from top to bottom
plt.gca().invert_yaxis()  

plt.title('Top 20 Companies by Revenue in 1971',fontsize=18)
plt.xlabel('Revenue (in millions)',fontsize=14)
plt.ylabel('Companies',fontsize= 15)

plt.show()
```

### 6.1 Plot Explanation

This Horizontal Bar Plot represents the Revenue (in millions) of the Top 20 companies by Revenue in 1971. The decision to choose this visualization was because the companies are easy to read and compare the difference in Revenue between different companies. The rationale for the monochrome plot is that the companies are not related and the plot is specifically comparing the Revenue in a particular year. The Horizontal Bar Plot allows a comparative analysis of the Revenue for each Company.

### 6.2 Plot Description

The plot shows the Top 20 Companies by Revenue in 1971, where the companies are ranked from highest Revenue to lowest Revenue, with General Motors at the top and General Telephone & Electronics at the bottom.

An interest point is that two out of the Top Five Companies are automotive, General Motors and Ford Motor. This reflects the dominance of the automotive industry in the 70s.

There are other companies with high Revenue represented in the ranking in 1971:

 * Oil and Gas Industries (Exxon Mobil, Gulf Oil, Texaco, Chevron, Mobil),
 * Steel Company (US Steel)
 * Electrical Equipments (General Electric, Westinghouse Electric)
 * IT Companies (IBM, ITT)
 
### 6.3 Plot Conclusion

The Horizontal Bar Plot shows that General Motors, Exxon Mobil, Ford Motor, General Electric and IMB have around 4.5 times more revenue compared to the last companies in the year 1971.

# 7. Top 10 Company's profit increased the most between 1990-1999

* **Replaced 'N.A.' with 0 in 'Profit' column to convert Dtype from 'object' to 'float' in the below cell**

```python
df['Profit (in millions)'] = df['Profit (in millions)'].replace('N.A.', 0)
```

* **Converted Dtype from 'object' to 'float' to ensure the plot can be visualised**


```python
df['Profit (in millions)'] = df['Profit (in millions)'].astype(float)
df['Profit (in millions)'].info()
```

* **Profit was grouped by Decade and then each group was aggregated with sum(), summing up the values in Profit column.**

```python
# Profit figures Grouped by Decade

profit_by_decades = df.groupby('Decade')['Profit (in millions)'].sum().reset_index()
profit_by_decades
```

```python
# Top 10 Profit figures by Company between 1990 and 1999

decade_90 = df[df['Decade'] == '1990 - 1999']

top10_decade_90 = decade_90.groupby('Company')['Profit (in millions)'].sum().sort_values(
    ascending=False).reset_index().head(10)

top10_decade_90
```

```python
# Bar Plot

# Plot size
plt.figure(figsize=(12, 7))
# Executing the Plot
ax = sns.barplot(data=top10_decade_90,x='Company', y='Profit (in millions)',
            palette='viridis')

plt.title('Top 10 Companies by Profit in the 90s',fontsize=20)
plt.xlabel('Companies',fontsize=16)
plt.ylabel('Profit (in millions)',fontsize=16)
plt.xticks(rotation=45)

# The x-axis labels were inverted to have the highest revenue at the top
ax.invert_xaxis()

plt.show()
```

### 7.1. Plot Explanation

This Bar Plot represents the Profits (in millions) of the Top 10 Company's whose Profit increased the most between 1990-1999.
The reason for choosing gradient colours was to identify the companies that have significatly increased their Profits, with the darkest colous assigned to the Companies with the highest Profits, while the lightest colours assigned to the Companies with least Profits between 1990-1999. By poltting a Bar Plot, it is clear and easy to visualize the distribution of Profits among the Companies.

### 7.2. Plot Description

Exxon Mobil and General Electic are the most profitable Companies, with Profits of over 55 million each, followed by Altria Group and Ford Motor, with Profits of over $40 million each.

It is worth mentioning that Exxon Mobil and Mobil Company are Oil and Gas industries, where Exxon Mobil is at the fist place, with Profits of 58 million, while Mobil is the least Profitable Company in the list with Profits of $21 million.

It is curious to observe that Intel (IT company) and Merck (healthcare) are not related industries, however both have Profits of nearly 30 million.

### 7.3. Plot Conclusion

A wide variety of industries have been analysed, where Exxon Mobil (Gas and Oil) is the most profitable between 1990-1999, which makes sense with the surge of the automotive industry in the 90s.

# 8. Average Revenue By Decade

```python
# Revenue figures by Decade
average_revenue = df.groupby('Decade')['Revenue (in millions)'].mean().reset_index()
average_revenue
```

```python
# Bar Plot

# Revenue grouped by Average
average_revenue = df.groupby('Decade')['Revenue (in millions)'].mean()

# Plot size
plt.figure(figsize=(12, 7))
# Executing the Plot
sns.barplot(x=average_revenue.index, y=average_revenue, palette='rocket_r')

plt.title('Average Revenue by Decade',fontsize=20)
plt.xlabel('Decades',fontsize=16)
plt.ylabel('Average Revenue (in millions)',fontsize=14)

plt.show()
```

### 8.1. Plot Explanation

This Bar Plot depicts the Average Revenue by Decade starting from 1950-1959 to 2000-2009. 
The reason for choosing gradient colours was to identify the Decades with the highest Average Revenue, with the darkest colous assigned to the Decades with the highest Average Revenue, while the lightest colours assigned to the Decades with the least Average Revenue. By plotting a Bar Plot, it is clear and easy to visualized the distribution of the Average Revenue by Decade.

### 8.2. Plot Description

It is evident that as the Decades pass, the Average Revenue increases. In the final Decade of the 20th century (1990-1999) was characterized by technological advancements and globalization with an Average Revenue of over 7 million, while in the first decade of the 21st century (2000-2009) was related to the continued global economic growth, with an Average Revenue of over 14000 million. 

### 8.3. Plot Conclusion

With the technological advancements continued global economic growth, the Decade 2000-2009 has incerased its Average Revenue by 100 times compared to the previous Decade and by 4200 times from the Decade 1950-1959 to 2000-2009.
