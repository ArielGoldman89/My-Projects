# Scania Trucks Analysis

The data presented in this dataset was obtained from Scania Trucks that are frequently used for heavy-duty transportation. Specifically, the dataset focuses on the performance of the Air Pressure System (APS), which generates pressurized air used for various truck functions like braking and gear changes. 
The dataset provides two classes: the positive class, which consists of component failures in a specific APS component, and the negative class, which includes trucks with failures that are not related to the APS system.

```python
# Filter columns with more than 70% missing values

missing_cols = df_failure.columns[df_failure.isna().sum() / len(df_failure) > 0.70]
print('There are {} columns with NaN above 70%.'.format(len(missing_cols)))
print('The missing columns are:',missing_cols)
```

```python
# drop columns with more than 70% missing values

df_failure = df_failure.drop(missing_cols, axis=1)
print('There are {} columns after dropping +70% missing values in each one.'.format(df_failure.shape[1]))
```

```python
# Replace categorical data "neg" and "pos" with 0 and 1

df_failure['class'].replace('neg', 0, inplace = True)
df_failure['class'].replace('pos', 1, inplace = True)
```

```python
# Calculate the mean 
mean_values = df_failure.mean(axis=0)

# Fill NaN values with the mean 
df_failure.fillna(mean_values, axis=0, inplace=True)
```

# Statistical Techniques

### Calculate the central tendency metrics of the APS component failures in both classes.

```python
df_failure['class'].unique()
```

```python
df_failure['class'].value_counts()
```

```python
# The 'class' column in the DF contains binary labels.

# ---> 1 represents the positive class
# ---> 0 represents the negative class.

print('The number of positive class :', df_failure['class'].value_counts()[1])
print('The number of negative class :', df_failure['class'].value_counts()[0])
```

### Central Tendency Metrics of the APS component failures in both classes


```python
# Mean and Median for positive class (class=1)
positive_class = df_failure[df_failure["class"] == 1]
pos_mean = positive_class.mean().mean()
pos_median = positive_class.median().median()

# Mean and Median for negative class (class=0)
negative_class = df_failure[df_failure["class"] == 0]
neg_mean = negative_class.mean().mean()
neg_median = negative_class.median().median()

# Mode for positive class (class=1)
positive_class = df_failure[df_failure["class"] == 1]
pos_mode = positive_class.mode()
# Mode for negative class (class=0)
negative_class = df_failure[df_failure["class"] == 0]
neg_mode = positive_class.mode()

print(f'Positive class mean: {pos_mean:.4f}')
print(f'Positive class median: {pos_median:.4f}')

print(f'\nNegative class mean: {neg_mean:.4f}')
print('Negative class median:', neg_median)

print('\nPositive class Mode:\n', pos_mode)
print('\nNegative class Mode:\n', neg_mode)
```

### VAR and STD of the APS component failures in both classes

```python
# Variance and STD for positive class (class=1)
positive_class = df_failure[df_failure["class"] == 1]
pos_var = positive_class.var().var()
pos_std = positive_class.std().std()

# Variance and STD for negative class (class=0)
negative_class = df_failure[df_failure["class"] == 0]
neg_var = negative_class.var().var()
neg_std = negative_class.std().std()

print('Positive class variance:', pos_var)
print('Positive class STD:', pos_std)

print('\nNegative class variance:', neg_var)
print('Negative class STD:', neg_std)
```

### Histogram to visualise the distribution of APS in positive class

```python
pos_data = df_failure[df_failure['class'] == 1].iloc[:, 1:]
                 
plt.hist(pos_data.values.flatten(), bins=10, alpha=0.5)
plt.xlabel("APS component failures")
plt.ylabel("Positive Class")
plt.title("Distribution APS component failures positive class")
plt.show()                      
```

### Box Plot to visualise the distribution of APS in positive class

```python
pos_data = df_failure[df_failure['class'] == 1].iloc[:, 1:]

plt.boxplot(pos_data.values, vert=False)
plt.yticks(range(1, pos_data.shape[1]+1), pos_data.columns)
plt.xlabel("Feature Values")
plt.title("Distribution of Features in Positive Class")
plt.show()
```

```python
pos_data = df_failure[df_failure['class'] == 1].iloc[:, 1:]

# Interquartile Range method
Q1 = pos_data.quantile(0.25)
Q3 = pos_data.quantile(0.75)
IQR = Q3 - Q1

# Cleaning up the outliers (values) of the Box Plot
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
pos_data = pos_data[~((pos_data < lower_bound) |(pos_data > upper_bound)).any(axis=1)]

print('There are {} remaining observations'.format(pos_data.shape[0]))
```

# Descriptive Statistical Analysis

### Binomial and Poisson Distribution

```python
df_failure.describe(include='all')
```

```python
# Percentage of positive and negative class

percentage_neg = (df_failure['class'].value_counts()[1] / len(df_failure)) * 100
print(f'Percentage of positive class {percentage_neg:.4f}')

percentage_pos = (df_failure['class'].value_counts()[0] / len(df_failure)) * 100
print(f'Percentage of negative class {percentage_pos:.4f}')
```

```python
from scipy.stats import binom
```

```python
# n = 50 (the company performs 50 truck inspections per week)
# K = 10 (the number we are looking for is at least 10 APS component failures)
# p = 0.016667 (which represents the probability of a component failing in a given inspection for positive class)
#    ---> the number calculated (pos class/total class = 1000/60,000 = 0.01667 )

prob_b = 1 - binom.sf(k = 9, n = 50, p = 0.01667)
print('The probability of observing at least 10 APS component failures:', prob_b)
```

```python
from scipy.stats import poisson
```

```python
# k = 8 or more APS failures in a given period
# mu = 10 (average number of APS component failures in a given period)

prob_p = poisson.sf(k = 7, mu = 10)
print('The probability of observing 8 or more APS component failures:', prob_p)
```
